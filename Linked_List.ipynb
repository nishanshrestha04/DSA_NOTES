Node class with three instance variables: `next`, `prev`, and `info`. The `__init__` method sets the initial values of these instance variables to `None`, `None`, and `0`, respectively.

```python
class SLList:
    def __init__(self):
        self.first = None
        self.last = None

    def insert_beg(self):
        element = int(input("Enter element to insert: "))
        node = Node()
        node.info = element
        if self.first == None:
            self.first = self.last = node
            node.next = None
        else:
            #Your code
            node.next = self.first
            self.first = node

    def insert_end(self):
        element = int(input("Enter element to insert: "))
        node = Node()
        node.info = element
        if self.first == None:
            self.first = self.last = node
            node.next = None
        else:
            #Your code
            self.last.next = node
            node.next = None
            self.last = node

    def insert_spec_pos(self):
        element = int(input("Enter element to insert: "))
        node = Node()
        node.info = element
        pos = int(input("Enter position to insert: "))
        if self.first == None:
            self.first = self.last = node
            node.next = None
        else:
            #Your code
            temp = self.first
            for i in range(0, pos):
                temp = temp.next
            node.next = temp.next
            temp.next = node

    def delete_first(self):
        if self.first == None:
            print("List is empty")
        elif self.first == self.last:
            self.first = None
            self.last = None
        else:
            #Your code
            self.first = self.first.next

    def delete_end(self):
        if self.first == None:
            print("List is empty")
        elif self.first == self.last:
            self.first = None
            self.last = None
        else:
            #Your code
            temp = self.first
            while temp.next.next != 0:
                temp = temp.next
            temp.next = None

    def delete_spec_pos(self):
        if self.first == None:
            print("List is empty")
        elif self.first == self.last:
            self.first = None
            self.last = None
        else:
            #Your code
            temp = self.first
            pos = int(input('Enter the node to delete'))
            for i in range(0, pos):
                temp = temp.next
            temp.next = temp.next.next


    def search(self):
        element = int(input("Enter element to search: "))
        temp = self.first
        pos = 1
        while temp != None:
            if temp.info == element:
                print("Element found at position:", pos)
                break
                temp = temp.next
                pos += 1
            if temp == None:
                print("Element not found in the list")

    def display(self):
        if self.first == None:
            print("List is empty")
        else:
            temp = self.first
            while temp != None:
                print(temp.info, "  ", end='')
                temp = temp.next
                print("")
lst = SLList()

while True:
    print("1. Insert at beginning")
    print("2. Insert at end")
    print("3. Insert at specific position")
    print("4. Delete first node")
    print("5. Delete last node")
    print("6. Delete node at specific position")
    print("7. Search element")
    print("8. Display list")
    print("9. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 9:
        break
    elif choice == 1:
        lst.insert_beg()
    elif choice == 2:
        lst.insert_end()
    elif choice == 3:
        lst.insert_spec_pos()
    elif choice == 4:
        lst.delete_first()
    elif choice == 5:
        lst.delete_end()
    elif choice == 6:
        lst.delete_spec_pos()
    elif choice == 7:
        lst.search()
    elif choice == 8:
        lst.display()
    else:
        print("Wrong choice")
        print("Bye!")
```

```python
class SLList:
    def __init__(self):
        self.first = None
        self.last = None
        self.length = 0

    def insert(self, element, pos=None):
        node = Node()
        node.info = element
        if self.first is None:
            self.first = self.last = node
            node.next = None
        elif pos == 1:
            node.next = self.first
            self.first = node
        elif pos is None or pos == self.length + 1:
            self.last.next = node
            self.last = node
            node.next = None
        elif 1 < pos <= self.length:
            temp = self.first
            for i in range(1, pos - 1):
                temp = temp.next
            node.next = temp.next
            temp.next = node
        self.length += 1

    def delete(self, pos=None):
        if self.first is None:
            print("List is empty")
        elif pos == 1:
            self.first = self.first.next
            if self.first is None:
                self.last = None
            self.length -= 1
        elif pos == self.length:
            if self.first == self.last:
                self.first = self.last = None
            else:
                temp = self.first
                while temp.next != self.last:
                    temp = temp.next
                temp.next = None
                self.last = temp
            self.length -= 1
        elif 0 < pos < self.length:
            temp = self.first
            for i in range(1, pos - 1):
                temp = temp.next
            temp.next = temp.next.next
            self.length -= 1
            if pos == self.length:
                self.last = temp
        else:
            print("Invalid position")


    def search(self, element):
        temp = self.first
        pos = 1
        while temp is not None:
            if temp.info == element:
                print("Element found at position:", pos)
                break
            temp = temp.next
            pos += 1
        if temp is None:
            print("Element not found in the list")

    def display(self):
        if self.first is None:
            print("List is empty")
        else:
            temp = self.first
            while temp is not None:
                print(temp.info, "  ", end='')
                temp = temp.next
            print("")

lst = SLList()

while True:
    print("1. Insert at beginning")
    print("2. Insert at end")
    print("3. Insert at specific position")
    print("4. Delete first node")
    print("5. Delete last node")
    print("6. Delete node at specific position")
    print("7. Search element")
    print("8. Display list")
    print("9. Exit")
    choice = int(input("Enter your choice: "))
    if choice == 9:
        break
    elif choice == 1:
        element = int(input("Enter element to be inserted: "))
        lst.insert(element, 1)
    elif choice == 2:
        element = int(input("Enter element to be inserted: "))
        lst.insert(element)
    elif choice == 3:
        element = int(input("Enter element to be inserted: "))
        pos = int(input("Enter position at which element is to be inserted: "))
        lst.insert(element,pos)
    elif choice == 4:
        lst.delete(1)
    elif choice == 5:
        lst.delete(lst.length-1)
    elif choice == 6:
        pos = int(input("Enter position of element to be deleted: "))
        lst.delete(pos)
    elif choice == 7:
        element = int(input("Enter element to search: "))
        lst.search(element)
    elif choice == 8:
        lst.display()
    else:
        print("Wrong choice")
        print("Bye!")
```
